```python
records = []
rn = []
while True:
    print('1. Insert Student Data')
    print('2. Search Student Data')
    print('3. Delete Student Data')
    print('4. Sort Student Data')
    print("5. Display all Student's Data")
    print('6. EXIT')
    ch = int(input('Enter your choice: '))
    if ch == 1:
        while True:
            roll=int(input('\nEnter Student Roll: '))
            if rn.count(roll)==0:
                rn.append(roll)
                break
            else: 
                print ('\nEntered Roll Number ALREADY PRESENT. TRY AGAIN!')
        name=input('Enter Student Name: ')
        marks=int(input('Enter Student Marks: '))
        records.append([roll, name, marks])
        print('\nRECORD SUCCESSFULLY INSERTED!\n')
    elif ch == 2:
        r = int(input('\nEnter the roll number to be searched: '))
        flag = 0
        for i in range(0,len(records)):
            if records[i][0] == r:
                print('\nStudent Record Found! --> ', records[i], '\n')
                flag = 1
        if flag == 0:
            print('\nNO RECORD FOUND!\n')
    elif ch == 3:
        r = int(input('\nEnter the roll number to be deleted: '))
        flag = 0
        for i in range(0,len(records)):
            if records[i][0] == r:
                print('\nStudent Record -->', records[i])
                records.pop(i)
                print('Deleted Successfully\n')
                flag = 1
                break
        if flag == 0:
            print('\nNO RECORD FOUND!\n')
    elif ch == 4:
        records.sort()
        print('\nRECORDS SORTED SUCCESSFULLY!')
        print(records, '\n')
    elif ch == 5:
        print('\nSTUDENT RECORDS -->')
        print(records, '\n')
    elif ch == 6:
        print('\nEXITING....')
        break
    else:
        print('\nSELECT A VALID OPTION!\n')
        
    
```
